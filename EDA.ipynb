```python

mask = df.isna().any(axis=0)
df.isna().mean()[mask].sort_values().plot(
    kind='bar', figsize=(15, 4),
    title='Porcentaje de valores faltantes por columna',
    ylabel='Relación de valores perdidos por columna'
)    
```
